```python
df_cp=df.copy()
df_cp['affected']= df['nkill']+df['nwound']
df_cp['affected'].value_counts()
```

```python
df1 = df_cp.sort_values(by='affected', ascending=False)[:20]
```

```python
heat =df1.pivot_table(index='country_txt', columns='iyear', values='affected')
heat.fillna(0,inplace=True)
```

```python
sns.heatmap(heat, cmap='OrRd')
plt.xlabel('YEARS',fontsize=12)
plt.ylabel('Country',fontsize=12)
plt.title('Heatmap with respect to Country and Year displaying the attacked', fontsize=15 )
```

```python
plt.figure(figsize=(20,9))
sns.barplot(df['region_txt'].value_counts()[:10].index,df['region_txt'].value_counts()[:10].values, palette='Paired')
plt.title('N of Terrorist Activities by Region', fontsize=20)
plt.xlabel('Regions', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```
